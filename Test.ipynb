```python
from sklearn import datasets
import pandas as pd
from sklearn.model_selection import train_test_split
import numpy as np
from sklearn.ensemble import GradientBoostingRegressor as GBM
from sklearn.metrics import mean_absolute_error
from sklearn.preprocessing import MaxAbsScaler
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV
```

```python
#Будем использовать датасет digits и стандартный бустинг из sklearn
```

```python
digits = datasets.load_digits()
data = pd.DataFrame(digits.data, columns=digits.feature_names)
```

```python
data['class'] = digits.target
data
```

```python
X = digits.data
Y = digits.target
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state=42)
```

```python
#Прогоним поиск по сетке для данных до стандартизации и нормализации, получим некоторое качество модели
```

```python
param_grid = {'learning_rate' : np.arange(0.1, 1, 0.1),
              'n_estimators' : np.arange(10, 200, 10)}

model1 = GBM(criterion='mse')

grid = GridSearchCV(model1, param_grid, n_jobs=8)

grid.fit(X_train, Y_train)
pred = grid.predict(X_test)
```

```python
print(grid.best_params_)
print(grid.best_score_)
```
